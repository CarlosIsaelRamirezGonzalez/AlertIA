```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image

class_labels = {
    0: "Arma blanca",
    1: "Arma corta",
    2: "Arma larga",
    3: "Ataque de perro",
    4: "Choques",
    5: "Encañonamiento",
    6: "Forcejeos",
    7: "Golpe",
    8: "Incendio",
    9: "Normalidad",
    10: "Patada",
    11: "Persona herida",
}
```

```python

image_path = "Imagenes/Imagen Cuchillo.jpg"

# Carga la imagen y la convierte en un array numpy
img = image.load_img(image_path, target_size=(256, 256))
img_array = image.img_to_array(img)

# Normaliza la imagen al rango 0-1
img_array = img_array / 255.0

# Expande las dimensiones para que se ajuste al formato de entrada del modelo
img_array = np.expand_dims(img_array, axis=0)

# Muestra la imagen utilizando matplotlib
plt.imshow(img_array[0])
plt.axis('off')  # Desactiva los ejes
plt.show()

# Haz predicciones 
predictions = model.predict(img_array)

# Obtener la clase predicha
predicted_class = np.argmax(predictions, axis=1)

# Obtener la certeza de la clase predicha
certainty = predictions[0][predicted_class[0]]

# Imprime la clase predicha
print(f"La clase predicha es: {class_labels[predicted_class[0]]}")
print(f'predicted class -> {predicted_class[0]}')
print(f"Certeza -> {certainty:.2f}")
```

```python
import cv2

# Diccionario de clases (debe coincidir con las clases que usaste en el entrenamiento)
class_labels = {
    0: "Arma blanca",
    1: "Arma corta",
    2: "Arma larga",
    3: "Ataque de perro",
    4: "Choques",
    5: "Encañonamiento",
    6: "Forcejeos",
    7: "Golpe",
    8: "Incendio",
    9: "Normalidad",
    10: "Patada",
    11: "Persona herida",
}

# Inicializa la captura de video
cap = cv2.VideoCapture(1)

while True:
    # Captura frame-by-frame
    ret, frame = cap.read()

    if not ret:
        break

    # Preprocesa la imagen para que coincida con el formato del modelo
    img = cv2.resize(frame, (256, 256))
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Haz predicciones
    predictions = model.predict(img_array)

    # Obtener la clase predicha y su certeza
    predicted_class = np.argmax(predictions, axis=1)
    certainty = predictions[0][predicted_class[0]]

    # Define la etiqueta a mostrar
    label = "Normalidad"
    if certainty > 0.90:
        label = class_labels[predicted_class[0]]

    # Muestra la etiqueta en el frame
    cv2.putText(frame, f'{label} ({certainty:.2f})', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Muestra el frame
    cv2.imshow('frame', frame)

    # Salir con la tecla 'q'
    if cv2.waitKey(0) & 0xFF == ord('q'):
        break

# Cuando todo esté listo, libera la captura
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.models import load_model
import cv2


# Diccionario de clases (debe coincidir con las clases que usaste en el entrenamiento)
class_labels = {
    0: "Arma blanca",
    1: "Arma corta",
    2: "Arma larga",
    3: "Ataque de perro",
    4: "Choques",
    5: "Encañonamiento",
    6: "Forcejeos",
    7: "Golpe",
    8: "Incendio",
    9: "Normalidad",
    10: "Patada",
    11: "Persona herida",
}

# Iniciar la captura de video desde la cámara
cap = cv2.VideoCapture(0)  # Usa 0 para la cámara web por defecto. Cambia el número si tienes varias cámaras

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Redimensionar y normalizar el fotograma
    img_array = cv2.resize(frame, (256, 256))
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    # Hacer predicciones
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)
    
    # Obtener la clase predicha
    predicted_label = class_labels[predicted_class[0]]
    
    # Mostrar la clase predicha en el fotograma
    cv2.putText(frame, f'Prediccion: {predicted_label}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    # Mostrar el fotograma
    cv2.imshow('Video', frame)
    
    # Salir si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar la captura de video y cerrar todas las ventanas
cap.release()
cv2.destroyAllWindows()
```
