```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.models import load_model
import cv2


# Diccionario de clases (debe coincidir con las clases que usaste en el entrenamiento)
class_labels = {
    0: "Arma blanca",
    1: "Arma corta",
    2: "Arma larga",
    3: "Ataque de perro",
    4: "Choques",
    5: "Encañonamiento",
    6: "Forcejeos",
    7: "Golpe",
    8: "Incendio",
    9: "Normalidad",
    10: "Patada",
    11: "Persona herida",
}

# Iniciar la captura de video desde la cámara
cap = cv2.VideoCapture(0)  # Usa 0 para la cámara web por defecto. Cambia el número si tienes varias cámaras

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Redimensionar y normalizar el fotograma
    img_array = cv2.resize(frame, (256, 256))
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    # Hacer predicciones
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)
    
    # Obtener la clase predicha
    predicted_label = class_labels[predicted_class[0]]
    
    # Mostrar la clase predicha en el fotograma
    cv2.putText(frame, f'Prediccion: {predicted_label}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    # Mostrar el fotograma
    cv2.imshow('Video', frame)
    
    # Salir si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar la captura de video y cerrar todas las ventanas
cap.release()
cv2.destroyAllWindows()
```
